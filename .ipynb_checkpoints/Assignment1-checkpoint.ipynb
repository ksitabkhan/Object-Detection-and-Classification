```python
modelConfiguration = 'C:\\Users\khadija sitabkhan\Documents\Assignments\Case_Studies\yolov4-tiny.cfg'
modelWeights= 'C:\\Users\khadija sitabkhan\Documents\Assignments\Case_Studies\yolov4-tiny.weights'
```

```python
net = cv2.dnn.readNetFromDarknet(modelConfiguration,modelWeights)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU)
```

```python
path_name = "Car_1.png"
image = cv2.imread(path_name)
file_name = os.path.basename(path_name)
filename, ext = file_name.split(".")
h, w = image.shape[:2]
whT = 320
# create 4D blob
# blob = cv2.dnn.blobFromImage(image, 1/255.0, (whT, whT), swapRB=True, crop=False)
blob = cv2.dnn.blobFromImage(image, 1/255.0, (whT, whT), [0,0,0],1, crop=False)
```

```python
net.setInput(blob)

layerNames= net.getLayerNames()
# print(layerNames)
print(net.getUnconnectedOutLayers())
```

```python
outputNames = [layerNames[i
                          [0]-1] for i in net.getUnconnectedOutLayers()]
outputs = net.forward(outputNames)
print(outputs[0].shape)
print(outputs[1].shape)
```

```python
def findObjects(outputs,img):
    ht,wt, ct=img.shape
    bbox=[]
    classIds=[]
    conf = []
    
    for output in outputs:
        for detection in output:
            scores= detection[5:]
            classId= np.argmax(scores)
            confidence = scores[classId]
            if confidence> confidence_threshold:
                w,h = int(detection[2]*wt), int(detection[3]*ht)
                x,y = int((detection[0]*wt) - w/2), int((detection[1]*ht) - h/2)
               
                bbox.append([x,y,w,h])
                classIds.append(classId)
                conf.append(float(confidence))
    print(len(bbox))
    indices = cv2.dnn.NMSBoxes(bbox,conf,confidence_threshold,nmsThreshold)
    print("indices is ", indices)
    for i in indices:
        i=i[0]
        box = bbox[i]
        x,y,w,h = box[0],box[1],box[2],box[3]
        overlay = img.copy()
        cv2.rectangle(overlay,(x,y),(x+w,y+h),(255,0,255),2)
        image = cv2.addWeighted(overlay, 0.6, img, 0.4, 0)
        # now put the text (label: confidence %)
#         cv2.putText(image, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX,fontScale=font_scale, color=(0, 0, 0), thickness=thickness)
        cv2.putText(image,f'{classNames[classIds[i]].upper()} : {round(conf[i]*100,2)}%',(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,0.6,(255,0,255),2)
        cv2.imwrite(filename + "_yolo32." + ext, image)   
```
